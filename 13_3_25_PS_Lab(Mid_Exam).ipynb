```python
p_A = 2/5
p_B = 1/2
p_A_and_B = 1/5

p_at_least_one = p_A + p_B - p_A_and_B
p_exactly_one = p_A + p_B - 2 * p_A_and_B

print(f"P(Revanth is selected for at least one project): {p_at_least_one}")
print(f"P(Revanth is selected for exactly one project): {p_exactly_one}")
```

```python
import numpy as np

x_values = [-2, -1, 0, 1, 2, 3]
p_values = [0.1, 'k', 0.2, '2k', 0.3, 'k']

k = (1 - (0.1 + 0.2 + 0.3)) / (1 + 2 + 1)

p_actual = [0.1, k, 0.2, 2*k, 0.3, k]
E_X = sum(x * p for x, p in zip(x_values, p_actual))
E_X2 = sum(x**2 * p for x, p in zip(x_values, p_actual))
Var_X = E_X2 - E_X**2

print(f"Value of k: {k}")
print(f"Expected Value E[X]: {E_X}")
print(f"Expected Value E[X^2]: {E_X2}")
print(f"Variance Var(X): {Var_X}")
```

```python
import scipy.stats as stats

n = 8
p = 0.5

p_at_least_6_heads = 1 - sum(stats.binom.pmf(k, n, p) for k in range(6))
p_at_most_6_heads = sum(stats.binom.pmf(k, n, p) for k in range(7))

print(f"P(At least 6 heads): {p_at_least_6_heads}")
print(f"P(At most 6 heads): {p_at_most_6_heads}")
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

mean = 70
std_dev = 3

p_greater_76 = 1 - stats.norm.cdf(76, mean, std_dev)
p_between_67_73 = stats.norm.cdf(73, mean, std_dev) - stats.norm.cdf(67, mean, std_dev)

print(f"P(Height > 76 inches): {p_greater_76}")
print(f"P(63 <= height <= 73): {p_between_67_73}")

x = np.linspace(60, 80, 1000)
pdf = stats.norm.pdf(x, mean, std_dev)

plt.plot(x, pdf, 'b', label="Normal Distribution")
plt.fill_between(x, pdf, where=(x >= 76), color='green', alpha=0.5, label="P(Height > 76)")
plt.fill_between(x, pdf, where=(x >= 67) & (x <= 73), color='orange', alpha=0.5, label="P(67 ≤ Height ≤ 73)")

plt.xlabel("Height (in inches)")
plt.ylabel("Probability Density")
plt.title("Normal Distribution of Heights")
plt.legend()
plt.show()
```
