```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

data = np.random.normal(60, 12, 1000)
plt.hist(data, bins=30, density=True, alpha=0.6, color='b')

x = np.linspace(20, 100, 1000)
pdf = stats.norm.pdf(x, 60, 12)
plt.plot(x, pdf, 'r', linewidth=2)

plt.show()
```

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

mean = 30
std_dev = 5

p1 = 1 - stats.norm.cdf(19, mean, std_dev)
p2 = stats.norm.cdf(19, mean, std_dev)
p3 = stats.norm.cdf(35, mean, std_dev) - stats.norm.cdf(19, mean, std_dev)

x = np.linspace(10, 50, 1000)
pdf = stats.norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, 'b')

plt.fill_between(x, pdf, where=(x >= 19), color='red', alpha=0.5)
plt.fill_between(x, pdf, where=(x <= 19), color='green', alpha=0.5)
plt.fill_between(x, pdf, where=(x >= 19) & (x <= 35), color='yellow', alpha=0.5)

plt.show()

print(f"P(X >= 19): {p1}")
print(f"P(X <= 19): {p2}")
print(f"P(19 <= X <= 35): {p3}")
```
