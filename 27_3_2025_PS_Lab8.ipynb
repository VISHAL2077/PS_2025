```python
import matplotlib.pyplot as plt

calories = [2000, 2200, 2500, 1800, 2300, 2600, 2100, 1900, 2700, 3000]
exercise = [60, 50, 30, 90, 45, 20, 75, 95, 15, 10]
bmi = [25, 26, 28, 23, 27, 30, 24, 22, 31, 33]

plt.figure(figsize=(20, 8))
plt.subplot(1, 3, 1)
plt.scatter(calories, bmi, color='blue')
plt.xlabel('Daily Calorie Intake')
plt.ylabel('BMI')
plt.title('Calorie Intake vs. BMI')

plt.subplot(1, 3, 2)
plt.scatter(exercise, bmi, color='red')
plt.xlabel('Exercise Duration (minutes)')
plt.ylabel('BMI')
plt.title('Exercise Duration vs. BMI')

plt.subplot(1, 3, 3)
plt.scatter(calories, exercise, color='green')
plt.xlabel('Daily Calorie Intake')
plt.ylabel('Exercise Duration (minutes)')
plt.title('Calorie Intake vs. Exercise Duration')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.stats import pearsonr

corr_calories_bmi, _ = pearsonr(calories, bmi)
corr_exercise_bmi, _ = pearsonr(exercise, bmi)
corr_calories_exercise, _ = pearsonr(calories, exercise)

print(f"Pearson Correlation (Calorie Intake vs. BMI): {corr_calories_bmi:.2f}")
print(f"Pearson Correlation (Exercise Duration vs. BMI): {corr_exercise_bmi:.2f}")
print(f"Pearson Correlation (Calorie Intake vs. Exercise Duration): {corr_calories_exercise:.2f}")
```

```python
if corr_calories_bmi > 0:
    print("Higher calorie intake is associated with a higher BMI.")
else:
    print("No strong relationship between calorie intake and BMI.")

if corr_exercise_bmi < 0:
    print("Higher exercise duration is associated with a lower BMI, indicating a negative correlation.")
else:
    print("No strong inverse relationship between exercise duration and BMI.")

if corr_calories_exercise < 0:
    print("Higher calorie intake is associated with lower exercise duration, suggesting an inverse relationship.")
else:
    print("No strong relationship between calorie intake and exercise duration.")
```
